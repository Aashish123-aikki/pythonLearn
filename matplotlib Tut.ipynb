```python
from matplotlib import pyplot as m
```

```python
m.plot([1,2,3],[4,2,6])
m.title("example")
m.xlabel("roll")
m.ylabel("Marks")
```

```python
from matplotlib import style
style.use('ggplot')
m.plot([4,5,6],[7,8,9],label="first")
m.plot([1,2,3],[4,5,6],label="second")
m.legend()
```

```python
m.scatter([1,2,3],[8,9,10])
```

```python
student={"arun":100,"aashish":90,"ravi":80,"kkrav":92}
keys=list(student.keys())
values=list(student.values())
m.bar(keys,values,color="blue")
m.title("Class Performance")
m.xlabel("Names")
m.ylabel("Marks")
m.show()
```

```python
#horizontal bar plot
student={"arun":100,"aashish":90,"ravi":80,"kkrav":92}
keys=list(student.keys())
values=list(student.values())
m.barh(keys,values,color="blue")
m.title("Class Performance")
m.ylabel("Names")
m.xlabel("Marks")
m.show()
```
